# coding a single neuron

```python
inputs = [1.0, 2.0, 3.0, 2.5]
bias = 2
weights = [0.2, 0.8, -0.5, 1.0]

neuron_output = (inputs[0]*weights[0] + 
                 inputs[0]*weights[0] + 
                 inputs[0]*weights[0] + 
                 inputs[0]*weights[0] + 
                 bias)
print(neuron_output)
```

# coding a layer (three neurons)

```python
weights = [[0.3, 0.5, -0.7, 1],
           [0.6, -0.4, 0.1, 0.3],
           [0.5, -0.6, 0.9, 0.2]]
bias = [1, 2, 1]

inputs = [1, 2, 1, 2]

layer = [(inputs[0]*weights[0][0] +
          inputs[1]*weights[0][1] +
          inputs[2]*weights[0][2] +
          inputs[3]*weights[0][3] + bias[0]), #neuron 1
          (inputs[0]*weights[1][0] +
          inputs[1]*weights[1][1] +
          inputs[2]*weights[1][2] +
          inputs[3]*weights[1][3] + bias[1]), # neuron 2
          (inputs[0]*weights[2][0] +
          inputs[1]*weights[2][1] +
          inputs[2]*weights[2][2] +
          inputs[3]*weights[2][3] + bias[2])] # neuron 3

print(layer)
```

# coding a layer with loops

```python
weights = [[0.3, 0.5, -0.7, 1],
           [0.6, -0.4, 0.1, 0.3],
           [0.5, -0.6, 0.9, 0.2]]
bias = [1, 2, 1]

inputs = [1, 2, 1, 2]

layer_output = []

for neuron_weights, neuron_bias in zip(weights, bias):
    neuron_output = 0

    for input, neuron_weight in zip(inputs, neuron_weights):
        neuron_output += input*neuron_weight
    
    neuron_output += neuron_bias

    layer_output.append(neuron_output)

print(layer_output)
```
