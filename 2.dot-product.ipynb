# Dot Product

## Rule : for dot product, number of column in first matrix must be equal to number of row in second matrix

```python
import numpy as np

a1 = np.array([[1],
      [2],
      [3]]) # shape = [3, 1]
    
a2 = np.array([[1, 2, 3]]) # shape = [1, 3]
    
a3 = np.array([[1,2, 3], 
               [1, 2, 3],
               [1, 2, 3]]) # shape = [3, 3]

a1.shape, a2.shape, a3.shape

# here, dot product of a1 and a3 is not allowed but of a2 and a3 can be done.
            
```

```python
weights = np.array([[0.3, 0.5, -0.7, 1],
           [0.6, -0.4, 0.1, 0.3],
           [0.5, -0.6, 0.9, 0.2]])
bias = [1, 2, 1]

inputs = np.array([1, 2, 1, 2]) # this works because numpy consider ti dimension as (4,)

layer_output = np.dot(weights, inputs) + bias

print(inputs.shape, layer_output.shape)

layer_output
```

```python
weights = np.array([[0.3, 0.5, -0.7, 1],
           [0.6, -0.4, 0.1, 0.3],
           [0.5, -0.6, 0.9, 0.2]])
bias = [1, 2, 1]

inputs = np.array([[1, 2, 1, 2],
                  [1, 2, 1, 2],
                  [1, 2, 1, 2]])

layer_output = np.dot(inputs, weights.T) + bias

layer_output
```
